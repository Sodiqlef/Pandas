```python
    import pandas
    
    # Loading the data in a pandas dataframe
    df = pandas.read_csv(r"C:\Users\DELL\Downloads\TDI dataset.csv")
    df = pandas.DataFrame(df)
    df
```

```python
# Creating a Pandas Series from the age column
ageSeries = pandas.Series(df["age"])
ageSeries
```

```python
numberOfMissingValues = ageSeries.isna().sum() #Counts number of missing values
ageSeriesMean = ageSeries.mean() #Calculates the mean
ageSeriesMedian = ageSeries.median() #Calculates the median
print("The total number of missing values:",numberOfMissingValues)
print("The mean:",ageSeriesMean)
print("The Median", ageSeriesMedian)
```
