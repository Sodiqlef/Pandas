```python
# All modules and libraries
from matplotlib import pyplot

import pandas
import numpy
```

```python
df = pandas.read_csv(r'C:\Users\DELL\Downloads\My python analysis\cleanedDf.csv')
```

```python
df.head()
```

```python
#Drop the unnamed column
df = df.drop(columns=["Unnamed: 0"])
```

```python
df.head()
```

```python
# A bar chart that shows the survival rate
survivor_count = df["survived"].value_counts()[0]
x = ["survivors", "non-survivors"]
y = [df["survived"].value_counts()[0], df["survived"].value_counts()[1] ]
pyplot.figure(figsize=(5,3))
pyplot.bar(x,y,color='pink')
pyplot.ylabel("Count")
pyplot.ylabel("Status")
pyplot.title("Survival rate")
pyplot.show()
```

```python
#check survival based on class
survived = df[df["survived"] == 1].groupby("pclass")["survived"].size().sort_values(ascending=False)
x = survived.index.astype(str)
y = survived.values
pyplot.figure(figsize=(5,5))
pyplot.bar(x,y, color="pink")
pyplot.xlabel("pclass")
pyplot.ylabel("Survivor count")

for i,v in enumerate(survived):
    pyplot.text(i, v, str(v), ha="center", va="bottom")
pyplot.show()
```
